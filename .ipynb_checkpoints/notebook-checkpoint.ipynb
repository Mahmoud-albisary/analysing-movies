# What affects Revenues and Adjusted Revenues?

## How budget(Adjusted) affects Revenue(Adjusted)?

```python
plt.scatter(df4["budget_adj"], df4["revenue_adj"])
plt.xlabel("Adjusted Budjet")
plt.ylabel("Adjusted Revenue")
plt.title("Adj_Budjet and Adj_Revenue plot")
plt.grid()
```

There is no clear relation between Adjusted Budget and Adjusted Revenue

## How the release year affects the adjusted revenue?

to plot graphs including revenue and budget the movies that had zero revenue or zero budget were removed to apply a fair analysis to the data because 0 revenue or budget means incorrect data

```python
plt.bar(df2["release_year"], df2["revenue_adj"])
plt.xlabel("Year")
plt.ylabel("Adjusted Revenue")
plt.title("Year and Adj Revenue bar chart")
plt.grid()
```

## How release year affects the adjusted revenue?

```python
plt.bar(df2["release_year"], df2["revenue"])
plt.xlabel("Year")
plt.ylabel("Revenue")
plt.title("Year and Revenue bar chart")
plt.grid()
```

The revenue was at highest in recent years but that was because of reduction in the price of dollar and that was observed because the revenue_adj was at highest in 1977 and 2009 without showing a clear relation between old and recent years and the revenue_adj

## How Genres affects revenue (adjacent)?

### The graph below  shows the relation between different genres and revenue (adjusted)
So first we should get the revenues of all movies of each genre and then get their mean values and then plot the graph

```python
groupgen = df3.groupby("genres")["revenue_adj"].mean()
groupgen.plot(kind = "bar")
plt.xlabel("genres")
plt.ylabel("Adjusted Revenue")
plt.title("genres and Adj_Revenue bar chart")
plt.grid()
```

Animated movies have the highest mean of revenue_adj but forigen ones have the lowest mean of revenue_adj.

```python
df2["month"] = df["release_date"].dt.month
grouped = df2.groupby("month")["revenue_adj"].mean()
```
